```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
# Download tokenizer (if not already done)
try:
    nltk.download('punkt')
except Exception as e:
    print(f"Error downloading punkt: {e}")
```

```python
# Input text
text = "This is a test. This test is only a test."
```

```python
# Tokenize the text into words and convert to lowercase to catch case duplicates
words = word_tokenize(text.lower())
```

```python
# Filter out non-alphabetic tokens (e.g., punctuation)
words = [word for word in words if word.isalpha()]
```

```python
# Count word frequencies
word_counts = Counter(words)
print(word_counts)
```

```python
# Find duplicate words (count > 1)
duplicates = {word: count for word, count in word_counts.items() if count > 1}
```

```python
# Display result
print("Duplicate Words Found:")
for word, count in duplicates.items():
    print(f"'{word}': {count} times")
```
