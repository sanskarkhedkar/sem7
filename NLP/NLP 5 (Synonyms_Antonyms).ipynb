```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

```python
#Display synsets
wn.synsets('happy')
wn.synsets('dog', pos=wn.VERB)        #synset having perticuar PoS category
synsets = wn.synsets('bike')
print(synsets)
```

```python
#Display name, definitiion, example of synsets
for synset in synsets:
  print("synset name:", synset.name())
  print("synset definition:", synset.definition())
  print("synset examples:", synset.examples())
```

```python
# Hypernym & hyponym
bike = wn.synset("bike.n.01")
print(bike)
print("Hypernyms of bike:", bike.hypernyms())
print("Hyponyms of bike:", bike.hyponyms())
```

```python
s2=wn.synsets("hand")
print(s2)
```

```python
# Meronym & Holonym
hand = wn.synset('hand.n.01')
print("Part Meronyms of hand:", hand.part_meronyms())
print("Part Holonyms of hand:", hand.part_holonyms())
```

```python
from nltk.corpus import wordnet as wn

# Synonyms of all senses of a word
synonyms = []
for syn in wn.synsets("happy"):        # go through each sense of "happy"
    for lemma in syn.lemmas():         # get synonyms for that sense
        synonyms.append(lemma.name())

print("All synonyms of 'happy':", set(synonyms))

# Synonyms of a particular sense of a word
sense2 = wn.synset("happy.a.04")  # "well expressed and to the point"
print("\nSense definition:", sense2.definition())
print("Synonyms:", [lemma.name() for lemma in sense2.lemmas()])

# Antonyms of all senses of a word
antonyms = []
for syn in wn.synsets("happy"):   # Loop over all senses (synsets) of "happy"
    for lemma in syn.lemmas():    # Loop over all lemmas (words) in that synset
        if lemma.antonyms():      # Check if the lemma has antonyms defined
            antonyms.append(lemma.antonyms()[0].name())    # Take the first antonym

print("\nAntonyms of 'happy':", set(antonyms))
```
