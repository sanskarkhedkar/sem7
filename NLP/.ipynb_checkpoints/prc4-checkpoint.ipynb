```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
#Tokenization
sentence = "He is running on the track. I was running on the same track. I told him that I run but you go"
sent_tok = nltk.sent_tokenize(sentence)
print(sent_tok)
token1 = nltk.word_tokenize(sent_tok[0])
token2 = nltk.word_tokenize(sent_tok[1])
token3 = nltk.word_tokenize(sent_tok[2])
#tokens = token1 + token2 + token3
print("Tokens",token1)
print("Tokens",token2)
print("Tokens",token3)
```

```python
#PoS Tagging
pos_tag1 = nltk.pos_tag(token1)
pos_tag2 = nltk.pos_tag(token2)
pos_tag3 = nltk.pos_tag(token3)
print("Part of Speech Tags",pos_tag1)
print("Part of Speech Tags",pos_tag2)
print("Part of Speech Tags",pos_tag3)
```

```python
#Chunking OR Shalow Parsing
grammar = r"""
  PP: {<IN><DT>?<JJ.*>*<NN.*>+}
  VP: {<VB>|<VBZ><VBG>|<VBZ><VBN>|<VBP>|<VBD>}
"""

chunker = nltk.RegexpParser(grammar)  # This is where RegexpParser is used
tree1 = chunker.parse(pos_tag1)
tree2 = chunker.parse(pos_tag2)
tree3 = chunker.parse(pos_tag3)

print(chunker)

print(tree1)
print(tree2)
print(tree3)
```
