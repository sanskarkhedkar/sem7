```python
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
text = """Ich Heiße Sanskar, ich bin 21 jahre alt. ich studiere im MIT ADT Universität. Meine Freunde sind Vrushank, Ishan, Samarth, Pranav und Raj. Harish ist Imposter."""
```

```python
text_lower = text.lower()
print(text_lower)
```

```python
sent_tokens = sent_tokenize(text)
print(sent_tokens)
```

```python
tokens = word_tokenize(text)
print(tokens)
```

```python
stop_words = set(stopwords.words('german'))
print("All standard Stop words:", stop_words)
print("Number of standard Stop words:", len(stop_words))
filtered_tokens = [word for word in tokens if word not in stop_words]
print("Tokens after Stopword Removal:", filtered_tokens)
```

```python
from nltk import PorterStemmer
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in filtered_tokens]
print("Stemmed Tokens:", stemmed_tokens)
```

```python
from nltk import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(t) for t in tokens]
print(lemmas)
```

```python
text3 = """Natural Language Processing tasks include tokenization, stemming, and lemmatization.
It is a field of Artificial Intelligence. Natural Language Processing is used in chatbots, sentiment analysis, and machine translation."""
```

```python
# 1. Lowercasing
text_lower = text3.lower()
print("Lowercased Text:", text_lower)
```

```python
# 2. Sentence Tokenization using NLTK
sentences1 = sent_tokenize(text3)
print("Sentence Tokenization using NLTK:")
print("Sentences:", sentences1)
```

```python
# 3. Word Tokenization
tokens1 = word_tokenize(text3)
print("Tokens:", tokens1)
```

```python
# 4. Stopword Removal
stop_words1 = set(stopwords.words('english'))
print("All standard Stop words:", stop_words1)
print("Number of standard Stop words:", len(stop_words1))
filtered_tokens1 = [word for word in tokens1 if word not in stop_words1]
print("Tokens after Stopword Removal:", filtered_tokens1)
```

```python
# 5. Stemming
ps = PorterStemmer()
stemmed_tokens1 = [ps.stem(word) for word in filtered_tokens1]
print("Stemmed Tokens:", stemmed_tokens1)
```

```python
# 6. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmas1 = [lemmatizer.lemmatize(t) for t in tokens1]
print(lemmas1)
```
