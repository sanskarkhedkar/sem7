```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# Example sentences
token1 = nltk.word_tokenize("I have a red bike")
token2 = nltk.word_tokenize("The dog is running")
token3 = nltk.word_tokenize("He is eating an apple")
```

```python
#PoS Tagging
pos_tag1 = nltk.pos_tag(token1)
pos_tag2 = nltk.pos_tag(token2)
pos_tag3 = nltk.pos_tag(token3)
print("Part of Speech Tags",pos_tag1)
print("Part of Speech Tags",pos_tag2)
print("Part of Speech Tags",pos_tag3)
```

```python
#Chunking OR Shalow Parsing
grammar = r"""
  PP: {<IN><DT>?<JJ.*>*<NN.*>+}
  VP: {<VB>|<VBZ><VBG>|<VBZ><VBN>|<VBP>|<VBD>}
"""

chunker = nltk.RegexpParser(grammar)  # This is where RegexpParser is used
tree1 = chunker.parse(pos_tag1)
tree2 = chunker.parse(pos_tag2)
tree3 = chunker.parse(pos_tag3)

print(chunker)

print(tree1)
print(tree2)
print(tree3)
```
