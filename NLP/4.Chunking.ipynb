```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
```

```python
sentences = [
    "He will go to the market.",
    "They have given up the fight.",
    "She runs quickly in the morning.",
    "He is running a successful business.",
    "We ate a delicious meal."
]
```

```python
#Chunking OR Shalow Parsing
grammar = r"""
  PP: {<IN><DT>?<JJ.*>*<NN.*>+}
  VP: {<VB.*><DT>?<JJ.*>*<NN.*>+}
      {<VB.*><RB.*>}
"""
```

```python
for sentence in sentences:
    print(f"\nSentence: {sentence}")
    
    # Tokenization
    tokens = nltk.word_tokenize(sentence)
    print("Tokens:", tokens)
    
   # POS Tagging
    pos_tags = nltk.pos_tag(tokens)
    print("POS Tags:", pos_tags) 
    
   #Chunking
    chunker = nltk.RegexpParser(grammar)  # This is where RegexpParser is used
    tree = chunker.parse(pos_tags)

    print("Parse:",chunker)

    print("Tree:",tree)

```
