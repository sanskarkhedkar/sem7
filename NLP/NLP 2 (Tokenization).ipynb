```python
 import nltk
 from nltk.corpus import stopwords
 from nltk.stem import PorterStemmer
 from nltk.tokenize import sent_tokenize, word_tokenize
 from nltk import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
text = """React is library of JavaScript .basically react is frontend library which is use to run time reacting in user server side where it give runtime effects on the every components of the any web application."""
text_lower = text.lower()
print("Lowercased Text:", text_lower)
```

```python
sentences = sent_tokenize(text)
print("Sentence Tokenization using NLTK")
print("Sentences:", sentences)
```

```python
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

```python
stop_words = set(stopwords.words('english'))
print("All standard stop words:", stop_words)
print("Number of standard Stop words:", len(stop_words))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("Tokens after Stopword Removal:", filtered_tokens)
```

```python
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in filtered_tokens]
print("Stemmed Tokens:", stemmed_tokens)
```

```python
lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(t) for t in tokens]
print("Lemmatized Tokens:", lemmas)
```
