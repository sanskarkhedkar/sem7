```python
import numpy as np
```

```python
 # Creating a 1D array using a list
array_1d = np.array([1, 2, 3, 4, 5])
print("1D Array: \n", array_1d)
```

```python
# Creating a 2D array
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:\n", array_2d)
```

```python
# Array characteristics
print("\nArray Shape:", array_2d.shape)
print("Array Data Type:", array_2d.dtype)
print("Array Size:", array_2d.size)
```

```python
# Basic operations on arrays
array_sum = np.sum(array_1d)
print("\nSum of 1D Array elements:", array_sum)
array_mean = np.mean(array_2d)
print("Mean of 2D Array elements:", array_mean)
```

```python
 # Array creation using arange and reshape
array_arange = np.arange(10)
array_reshaped = array_arange.reshape(2, 5)
print("\nArray created using arange and reshaped: \n", array_reshaped)
```

```python
# Creating arrays of zeros and ones
array_zeros = np.zeros((3, 3))
print("\nArray of zeros: \n", array_zeros)
array_ones = np.ones((2, 4))
print("\nArray of ones: \n", array_ones)
```

```python
# Creating an identity matrix
identity_matrix = np.eye (4)
print("\nIdentity matrix:\n", identity_matrix)
```

```python
# Element-wise operations
array_squared = array_1d ** 2
print("\nElement-wise square of 1D Array: \n", array_squared)
array_sum_2d = np.sum(array_2d, axis=0) # Sum along columns
print("\nSum of 2D Array elements along columns: \n", array_sum_2d)
array_sum_2d_row= np. sum (array_2d, axis=1)
np.sum(array_2d, axis=1) # Sum along rows
print("Sum of 2D Array elements along rows: \n", array_sum_2d_row)
```

```python
# Array indexing and slicing
print("\nElement at index 1 in 1D Array:", array_1d[1])
print("Element at position (0, 2) in 2D Array: ", array_2d [0, 2])
array_slice_1d = array_1d[1:4]
print("\nSliced 1D Array (index 1 to 3): \n", array_slice_1d)
array_slice_2d = array_2d[:, 1:3] # Slicing columns 1 to 2
print("Sliced 2D Array (columns 1 to 2):\n", array_slice_2d)
```

```python
# Boolean indexing
array_bool = array_1d [array_1d > 3]
print("\nElements greater than 3 in 1D Array: \n", array_bool)
```

```python
# Operations on reshaped arrays
array_reshaped_T = array_reshaped. T # Transpose of reshaped array
print("\nTranspose of the reshaped array:\n", array_reshaped_T)
```

```python
# Flattening the array
array_flattened = array_2d.flatten()
print("\nFlattened 2D Array: \n", array_flattened)
```

```python
 # Concatenating arrays
array_concat = np.concatenate((array_1d, array_1d), axis=0)
print("\nConcatenated 1D Arrays along axis 0:\n", array_concat)
```

```python
 # Sorting an array
sorted_array = np.sort(array_1d)
print("\nSorted 1D Array:\n", sorted_array)
```

```python
 # Finding unique elements
unique_elements = np. unique (array_2d)
print("\nUnique elements in 2D Array: \n", unique_elements)
```

```python
 # Finding maximum and minimum
max_value = np.max(array_2d)
min_value = np.min(array_2d)
print("\nMaximum value in 2D Array:", max_value)
print("Minimum value in 2D Array:", min_value)
```
