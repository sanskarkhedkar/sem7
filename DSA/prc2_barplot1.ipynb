```python
import matplotlib.pyplot as plt
# Number of preferences for different libraries
library = ['Matplotlib', 'Seaborn', 'Plotly', 'Plotnine']
chosen_by = [2500, 1800, 3000, 2200]
```

```python
# Vertical Bar Plot
plt.bar(library, chosen_by, color='skyblue')
plt.xlabel('Visualization Library')
plt.ylabel('Number of Enthusiasts')
plt.title('Which Visualization Library Do People Prefer?')
plt.show()
```

```python
# Horizontal Bar Plot
plt.barh(library, chosen_by, color='lightgreen')
plt.xlabel('Visualization Library')
plt.ylabel('Number of Enthusiasts')
plt.title('Which Visualization Library Do People Prefer?')
plt.show()
```

```python
plt.bar(library, chosen_by, color='skyblue')
plt.xlabel('Visualization Library')
plt.ylabel('Number of Enthusiasts')
plt.title('Which Visualization Library Do People Prefer?')

# Adding annotations
for i, value in enumerate(chosen_by):
    plt.text(i, value + 5, str(value), ha='center')
plt.show()
```

```python
import numpy as np

# Define library names
library = ['Matplotlib', 'Seaborn', 'Plotly', 'Plotnine']

# Number of Enthusiasts for different regions
enthusiasts_north = [2000, 1500, 2500, 2000]
enthusiasts_south = [1500, 1300, 2000, 1800]
bar_width = 0.35
x = np.arange(len(library))

# Grouped Bar Plot
plt.bar(x - bar_width/2, enthusiasts_north, bar_width, label='North Region', color='skyblue')
plt.bar(x + bar_width/2, enthusiasts_south, bar_width, label='South Region', color='lightgreen')

# Adding labels and title
plt.xlabel('Visualization Library')
plt.ylabel('Number of Enthusiasts')
plt.title('Regional Preferences for Visualization Libraries (Grouped)')
plt.xticks(x, library)
plt.legend(title='Regions')
plt.show()
```

```python
# Define library names
library = ['Matplotlib', 'Seaborn', 'Plotly', 'Plotnine']

# Number of Enthusiasts for different regions
enthusiasts_north = [2000, 1500, 2500, 2000]
enthusiasts_south = [1500, 1300, 2000, 1800]
x = np.arange(len(library))

# Stacked Bar Plot
plt.bar(x, enthusiasts_north, label='North Region', color='skyblue')
plt.bar(x, enthusiasts_south, bottom=enthusiasts_north, label='South Region', color='lightcoral')

# Adding labels and title
plt.xlabel('Visualization Library')
plt.ylabel('Number of Enthusiasts')
plt.title('Regional Preferences for Visualization Libraries (Stacked)')
plt.legend()
plt.show()
```

```python
import pandas as pd
data = pd.read_csv("C:/Users/sansk/Downloads/kl.csv", encoding='latin1')
```

```python
data = data.drop(axis = 1, columns=['Photo','Flag', 'Club Logo'])  
data.head()
```

```python
def get_value(value):
    try:
        if isinstance(value, str):
            value = value.replace('', '')
            if 'M' in value:
                return float(value.replace('M', '')) * 1_000_000
            elif 'K' in value:
                return float(value.replace('K', '')) * 1_000
            else:
                return float(value)  # in case it's just a raw number like 100
        return 0.0  # for NaN or unexpected types
    except:
        return 0.0
```

```python
data['Value'] = data['Value'].apply(get_value)
data['Wage'] = data['Wage'].apply(get_value)
```

```python
print(data[['Value', 'Wage']].describe())
```

```python
by_nation = data.Nationality.value_counts()
top_10_nation = by_nation[:10]
top_10_nation
```

```python
top_10_nation.plot(kind='bar'); # The ';' is to avoid showing a message before the chart
```

```python
# We can also plot horizontally by using 'barh' in 'kind' argument
top_10_nation.plot(kind='barh');
```

```python
# Method 2: plt.bar() in matplotlib - we input x and y arguments
plt.bar(top_10_nation.index, top_10_nation);
```

```python
# Horizontally
plt.barh(top_10_nation.index, top_10_nation);
```
