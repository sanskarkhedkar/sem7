```
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4])
y=x*2

plt.plot(x,y)
plt.show()
```

```
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.figure(figsize=(8,6))

plt.plot(x,y,marker='o',linestyle='-')

for i,(xi,yi) in enumerate(zip(x,y)):
    plt.annotate(f'({xi},{yi})',(xi, yi), textcoords="offset points", xytext=(0,10), ha='center')

plt.title("Line chart with Annotate")
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.grid(True)
plt.show()
```

```
import numpy as np # useful for many scientific computing in Python
import pandas as pd # primary data structure library

df_can1 = pd.read_excel(
    'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2
)

print('Data read into a pandas dataframe!')
df_can1.head()
df_can1.shape
```

```
df_can1.info()
df_can1.columns.values
df_can1.index.values

print(type(df_can1.columns))
print(type(df_can1.index))

df_can1.columns.tolist()
df_can1.index.tolist()

print(type(df_can1.columns.tolist()))
print(type(df_can1.index.tolist()))

# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can1.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can1.head(2)

df_can1.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can1.columns
```

```
df_can1.dtypes
```

```
df_can1['Total'] = df_can1.apply(pd.to_numeric, errors='coerce').sum(axis=1)
```

```
df_can1.shape
```

```
df_can1.isnull().sum()
df_can1.describe()
```

```
df_can1[(df_can1['Continent']=='Asia') & (df_can1['Region']=='Southern Asia')]

print('data dimensions:', df_can1.shape)
print(df_can1.columns)
df_can1.head(2)
```

```
import matplotlib.pyplot as plt

# Define the years list as integers (not strings) since DataFrame columns are integers
years = list(range(1980, 2014))

# Set the index to Country if not already set
if 'Country' in df_can1.columns:
    df_can1.set_index('Country', inplace=True)

# Now get the Haiti data
haiti = df_can1.loc['Haiti', years]
haiti.head()

haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```
